```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.arange(0, 15, 0.1)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.title('sin(x) / cos(x)')
plt.xlabel('time')
plt.ylabel('amplitude')

plt.grid()

plt.xlim(0, 15)
plt.ylim(-1,1)

plt.plot(x, y1, label='sin(x)', color='r', ls='--')
plt.plot(x, y2, label='cos(x)', color='g', ls='--')

plt.legend(loc='best')

plt.show()
```
